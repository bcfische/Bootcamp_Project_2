```python
from bs4 import BeautifulSoup
import requests
```

```python
response = requests.get('https://www.imdb.com/list/ls066620113/?ref_=otl_3')
```

```python
soup = BeautifulSoup(response.text, 'lxml')
```

```python
results = soup.find_all('div', class_='list-description')
results
```
